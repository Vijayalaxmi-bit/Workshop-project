```python
# data visualization using matplotlib Library
import matplotlib.pyplot as plt
demo = data.groupby("year")["amount"].sum()
plt.plot(demo.index, demo.values,marker='o',markerfacecolor="Red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total Amount")
plt.title("year wise Amount investament",color="red",fontsize=16)
plt.grid(linestyle="--",color="skyblue",linewidth=1.5)
plt.xlabel("year",color="blue",fontsize=12)
plt.ylabel("Total Amount",color="blue",fontsize=12)
plt.legend()
plt.show()



```

```python
Demo = data["city"].value_counts().head(10)
color=["black","red","yellow","pink","green","purple","orange","brown","skyblue","blue"]
plt.bar(Demo.index,Demo.values,color=color,edgecolor="black")


plt.xticks(rotation=45)
plt.title("Top 10 startup company's in India")
plt.xlabel("City Names")
plt.ylabel("Number of start up")
plt.grid(True)
```

```python
demo = data["vertical"].value_counts().head(10)
color=["blue", "black", "red", "pink", "green", "purple", "brown", "skyblue", "orange", "lightgreen"]

plt.figure(figsize=(10,6))
plt.barh(demo.index[::-1], demo.values[::-1], color=color,alpha=0.6)
plt.xticks(rotation = 45)
plt.title("Top 10 Sectors startup in India",color="red",fontsize=20)
plt.xlabel("Number of startup copmany",color = "blue",fontsize=12)
plt.ylabel("Sectors",color = "red",fontsize=12)
plt.grid(axis = 'x',linestyle="--", color="black")
```

```python
Demo = data["round"].value_counts().head(7)
color=["blue", "black", "red", "pink", "green", "purple", "brown", "skyblue", "orange", "lightgreen"]

plt.figure(figsize=(10,6))
plt.pie(demo.values, labels=demo.index, colors=color, autopct="%1.1f%%")
plt.title("Rounds distributionin percentage ",color="red",fontsize=20)

```

```python
from matplotlib.lines import lineStyles
#  The relationship between funding year and funding Amount

plt.figure(figsize=(10,6))
plt.scatter(data["year"],data["amount"],color = "red")
plt.title("Funding amount by year")
plt.xlabel("Funding Year")
plt.ylabel("Funding Amount")
plt.grid(linestyle="--")
plt.show()
```

```python
# which funding round occurs most frequently
import seaborn as sns
Demo = data["round"].value_counts().head(10).index
plt.figure(figsize=(10,6))
sns.countplot(data,y="round",order=Demo,palette="viridis")
plt.title("Most frequently funding round")
plt.xlabel("Number of funding")
plt.ylabel("Funding Round")
plt.show()


```
